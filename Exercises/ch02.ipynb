## 14

> Define a function `supergloss(s)` that takes a synset `s` as its argument and returns a string consisting of the concatenation of the definition of `s`, and the definitions of all the hypernyms and hyponyms of `s`.

```python
def supergloss(s):
    ret = s.definition() + '. '
    for hypernym in s.hypernyms():
        ret = ret + hypernym.definition() + '. '
    for hyponym in s.hyponyms():
        ret = ret + hyponym.definition() + '. '
    return ret
```

```python
supergloss(wn.synset('car.n.01'))
```

## 17

> Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
def freq_words(text, n=50):
    fd = nltk.FreqDist(text)
    stopwords = nltk.corpus.stopwords.words('english')
    return sorted([word for word in fd if word not in stopwords], key=lambda x: fd[x], reverse=True)[:n]
```

```python
print(freq_words(brown.words(categories='news')))
```

## 18

> Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def freq_bigrams(text, n=50):
    fd = nltk.FreqDist(nltk.bigrams(text))
    stopwords = nltk.corpus.stopwords.words('english')
    return sorted([bigram for bigram in fd if bigram[0] not in stopwords and bigram[1] not in stopwords],
                  key=lambda x: fd[x], reverse=True)[:n]
```

```python
print(freq_bigrams(brown.words(categories='news')))
```

## 21

> Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary.

```python
def count_syllables(text):
    entries = nltk.corpus.cmudict.entries()
    entries_dict = dict(entries)
    ret = 0
    for word in text:
        ret += len(entries_dict.get(word.lower(), []))
    return ret
```

```python
count_syllables(brown.words(categories='news'))
```

## 22

> Define a function `hedge(text)` that processes a text and produces a new version with the word `'like'` between every third word.

```python
def hedge(text):
    ret = []
    for i in range(len(text)):
        ret.append(text[i])
        if i % 2 == 1:
            ret.append('like')
    return ret
```

```python
print(hedge(brown.words(categories='news'))[:20])
```

> **Zipf’s Law**: Let $f(w)$ be the frequency of a word $w$ in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf’s Law states that the frequency of a word type is inversely proportional to its rank (i.e., $f × r = k$, for some constant $k$). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

> a. Write a function to process a large text and plot word frequency against word rank using `pylab.plot`. Do you confirm Zipf’s law? (Hint: it helps to use a logarithmic scale.) What is going on at the extreme ends of the plotted line?

> b. Generate random text, e.g., using `random.choice("abcdefg ")`, taking care to include the space character. You will need to `import random` first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, generate the Zipf plot as before, and compare the two plots. What do you make of Zipf’s Law in the light of this?

```python
import matplotlib.pyplot as plt
from numpy import log
```

```python
def plot_word_freq(text):
    fd = nltk.FreqDist(text)
    words = sorted(list(fd.keys()), key=lambda x: fd[x], reverse=True)
    x = [log(i) for i in range(1, len(words) + 1)]
    y = [log(fd[key]) for key in words]
    plt.plot(x, y)
    plt.show()
```

```python
plot_word_freq(brown.words(categories='mystery'))
```

From the figure above we can see that Zipf's Law holds.

> Define a function `find_language()` that takes a string as its argument and returns a list of languages that have that string as a word. Use the `udhr` corpus and limit your searches to files in the Latin-1 encoding.

```python
from nltk.corpus import udhr
```

```python
def find_language(text):
    ret = []
    for language in udhr.fileids():
        if language.endswith('-Latin1') and text in udhr.words(language):
            ret.append(language)
    return ret
```

```python
find_language('aa')
```
